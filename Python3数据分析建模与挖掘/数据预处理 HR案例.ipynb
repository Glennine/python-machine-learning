```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import SelectKBest,RFE,SelectFromModel
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import Normalizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split#用于拆分训练集和测试集
```

```python
'''
s1:satisfaction_level---False:MinMaxScaler,True:StandardScaler
le:last_evaluation--False:MinMaxScaler,True:StandardScaler
npr:number_project--False:MinMaxScaler,True:StandardScaler
amh:average_monthly_hours--False:MinMaxScaler,True:StandardScaler
一般来说，数字比较大的就可以采取LabelEncoder,OneHotEncoder，
分装对数据进行一些属性丢失处理
tsc:time_spend_company --False:MinMaxScaler,True:StandardScaler
离散值的分装处理
ss:sales--False:LabelEnconder,True:HotEncoder
sa:salary--False:LabelEnconder,True:HotEncoder
'''
def hr_preprocessing(s1=False, le=False,npr=False,amh=False,\
                     tsc=False,ss=False,sa=False,lower_d=False,ld_n=1):
    df = pd.read_csv('d:\HR.csv')
    #2.得到标注
    label = df['left']
    df = df.drop('left',axis=1)#如果不写1的话，就会默认以行删除，而这里需要删除这一列
    #1.清洗数据
    df = df.dropna(subset=['satisfaction_level','last_evaluation'])
    df = df[df['satisfaction_level']<=1][df['salary']!='nme']
    #3.特征选择(一般对于那些不重要的，相关度低的特征可以舍弃)
    #4.特征处理
    scaler_lst = [s1, le, npr, amh, tsc]#这里是选择使用归一化F/标准化T
    column_lst = ['satisfaction_level','last_evaluation','number_project',\
                  'average_montly_hours','time_spend_company']
    for i in range(len(scaler_lst)):
        if not scaler_lst[i]:
            df[column_lst[i]]=\
            MinMaxScaler().fit_transform(df[column_lst[i]].values.reshape(-1,1)).reshape(1,-1)[0]
        else:
            df[column_lst[i]]=\
            StandardScaler().fit_transform(df[column_lst[i]].values.reshape(-1,1)).reshape(1,-1)[0]
    scaler2_lst = [ss,sa]
    column2_lst = ['sales','salary']
    for i in range(len(scaler2_lst)):
        if not scaler2_lst[i]:
            if column2_lst[i]=="salary":
                df[column2_lst[i]]=[map_salary(s) for s in df['salary'].values]
            else:
                df[column2_lst[i]]=LabelEncoder().fit_transform(df[column_lst[i]])
                df[column2_lst[i]]=\
            MinMaxScaler().fit_transform(df[column2_lst[i]].values.reshape(-1,1)).reshape(1,-1)[0]
        else:
            df = pd.get_dummies(df,columns=[column2_lst[i]])
    #5.降维问题
    if lower_d:
        return PCA(n_components=ld_n).fit_transform(df.values)
    #LinearDiscriminantAnalysis(n_components=ld_n)LDA是受限制的，所以可以先考虑PCA
    return df

d = dict([("low",0),("medium",1),("high",2)])
def map_salary(s):
    return d.get(s,0)

def main():
    print(hr_preprocessing(s1=True,le=True,ss=True))

if __name__ == '__main__':
    main()
```
